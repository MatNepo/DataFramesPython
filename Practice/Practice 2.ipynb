### Задание 1 *

Напишите функцию, которая может выполнять арифметические операции над pd.Series. Первые два параметра функции - pd.Series, над которыми проводится операция, третий параметр может принимать следующие значения: 'add' -сложение, 'subtract' - вычитание, 'multiply' - умножение, 'divide' - деление.

```python
def calculate_basic_operations(
    a: pd.Series,
    b: pd.Series,
    operation: list
) -> pd.Series:
    # создадим словарь с командами
    operations = {
        'add': a + b,
        'subtract': a - b,
        'multiply': a * b,
        'divide': a / b
    }

    # Проверим, что введённая операция существует
    if operation not in operations:
        raise ValueError("Неподдерживаемая операция. Поддерживаемые операции: 'add', 'subtract', 'multiply', 'divide'")

    return operations[operation]
```

```python
ds1 = pd.Series([2, 4, 6, 8, 10])
ds2 = pd.Series([1, 3, 5, 7, 9])
calculate_basic_operations(ds1, ds2, 'add')
```

Напишите функцию, которая будет возвращать строку, содержащую значения среднего арифметического и стандартного отклонения pd.Series в формате '{mean} +/- {std}'

```python
def make_mean_std_string(
    input_series: pd.Series,
) -> str:
    mean = input_series.mean()
    std = input_series.std()
    result_string = f'{mean:.2f} +/- {std:.2f}'
    return(result_string)
```

Напишите функцию, которая на вход принимает произвольный датафрейм, на выходе функции должен быть словарь, который содержит индексы максимальных числовых значений в качестве значений и названия столбцов в качестве ключей. Функция должна преобразовывать столбцы в числовые значения, если столбец содержит более 50% пропущенных значений, то индексы максимальных значений по нему не должны возвращаться.

```python
df_tasks = pd.DataFrame(
    {
        'task_num': list(range(6)),
        'asignee': ['Любовь', 'Евегений', 'Сергей', 'Светлана', 'Николай', 'Борис'],
        'in_progress': [True, False, True, True, True, False],
        'time': [10, 5, 2, 7, 4, 1],
    },
)
```

```python
def find_max_indexes(df: pd.DataFrame) -> dict:
    result_dict = {}

    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            # Преобразование столбца в числовые значения
            df[column] = pd.to_numeric(df[column], errors='coerce')

            # Проверка наличия более 50% пропущенных значений
            if df[column].isna().mean() <= 0.5:
                # Находим индекс максимального значения в столбце
                max_index = df[column].idxmax()
                result_dict[column] = max_index

    return result_dict
```

```python
find_max_indexes(df_tasks)
```

Напишите функцию, которая принимает на вход датафрейм, похожий по своей структуре на df_tasks. Среди всех значений необходимо вывести те, для которых in_progress = True, при этом значение time укладывается в межквартильный интервал по столбцу time, который расчитывается на все датасете (межквартильный интервал - значения между 25-м и 75-м процентилем). На выходе функции должен быть датафрейм, содержащий такие значения.

```python
def find_inrange_assignees(df: pd.DataFrame) -> pd.DataFrame:
    # Фильтровать строки, где in_progress имеет значение True
    in_progress_df = df[df['in_progress']]

    # Рассчитайте межквартильный размах для столбца «время»
    q25 = np.percentile(df['time'], 25)
    q75 = np.percentile(df['time'], 75)
    iqr = q75 - q25

    # Фильтровать строки, в которых «время» находится в межквартильном диапазоне
    in_range_df = in_progress_df[(in_progress_df['time'] >= q25) & (in_progress_df['time'] <= q75)]

    return in_range_df
```

### Задание №25 *

Напишите функцию, которая будет возвращать копию датафрейма без первых или последних (в зависимости от значения параметра drop_last) records_num записей. На вход функция должна принимать датафрейм, параметр records_num, отвечающий за количество записей для удаления и булевый параметр drop_last, который в случае, если он True функция убирает последние записи, иначе функция убирает первые записи. Если количество записей в датафрейме меньше records_num - нужно вывести копию исходного датафрейма.

```python
def drop_records(
    df: pd.DataFrame,
    records_num: int,
    drop_last: bool = True,
) -> pd.DataFrame:
    if len(df) <= records_num:
        # Если количество записей меньше или равно records_num, возвращаем копию исходного датафрейма
        return df.copy()

    if drop_last:
        # Удаление последних records_num записей
        result_df = df.iloc[:-records_num].copy()
    else:
        # Удаление первых records_num записей
        result_df = df.iloc[records_num:].copy()

    return result_df
```

```python
# Пример использования:
# Замените df_example на ваш реальный датафрейм
df_example = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'value': [10, 20, 30, 40, 50]
})

# Удалить первые две записи
result_df = drop_records(df_example, records_num=2, drop_last=False)
result_df
```

```python
# Удалить последние две записи
result_df = drop_records(df_example, records_num=2, drop_last=True)
result_df
```

### Задание №15*

Создайте новое поле full_price, в котором содержится результат произведения Quantity и Price. Затем получите столбец full_price_houndreds, содержайщий результат целочисленного деления на 100 full_price, приведите его к int. Значения full_price ниже  0 замените на -1 в full_price_houndreds.

```python
# Создание нового столбца 'full_price' как произведение Quantity и Price
df_retail['full_price'] = df_retail['Quantity'] * df_retail['Price']

# Создание нового столбца 'full_price_hundreds' как целочисленное деление 'full_price' на 100
df_retail['full_price_hundreds'] = (df_retail['full_price'] // 100).astype(int)

# Замена значений 'full_price_hundreds' меньше 0 на -1
df_retail.loc[df_retail['full_price_hundreds'] < 0, 'full_price_hundreds'] = -1

# Вывод только строк, где full_price_hundreds = -1
result_df = df_retail[df_retail['full_price_hundreds'] == -1]

# Вывод первых нескольких строк для проверки
result_df[['Description', 'full_price', 'full_price_hundreds']].head()
```

### Задание №16*

Создайте переменную work_hours, в которой будут содержаться значения True, если InvoiceDate приходился на день с понедельника по пятницу с 9 до 18 часов не включая последний.

```python
# Создание переменной work_hours
df_retail['work_hours'] = ((df_retail['InvoiceDate'].dt.dayofweek < 5) &  # Пн-Пт
                            (df_retail['InvoiceDate'].dt.hour >= 9) &
                            (df_retail['InvoiceDate'].dt.hour < 18))

# Вывод первых нескольких строк для проверки
df_retail[['InvoiceDate', 'work_hours']].head(20)
```

### Задание №17*

Найдите все товары, в описании которых есть слова 'CHRISTMAS', 'SANTA' и 'REINDEER'. Среди этих товаров определите месяц и день InvoiceDate по моде и выведите его при помощи стандартного вывода в текстовом формате без плавающей точки.

* case=False - используется для игнорирования регистра при поиске слов в описании
* na=False - используется для исключения пропущенных значений

```python
from calendar import month_name

# Фильтрация товаров с заданными словами в описании
filtered_df = df_retail[df_retail['Description'].str.contains('CHRISTMAS|SANTA|REINDEER', case=False, na=False)]

# Определение месяца и дня InvoiceDate по моде
mode_month = filtered_df['InvoiceDate'].dt.month.mode().values[0]
mode_day = filtered_df['InvoiceDate'].dt.day.mode().values[0]

# Вывод месяца и дня без плавающей точки
print(f"Month: {int(mode_month)} ({month_name[mode_month]})\nDay: {int(mode_day)}")
```

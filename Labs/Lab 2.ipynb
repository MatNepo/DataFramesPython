### Задание №2

**Условие**: Отсортируйте датафрейм в порядке возрастания года (year).

Чтобы отсортировать датафрейм по определённому значению воспользуемся методом `.sort_values()`, который выглядит следующим образом:
```
DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)
```
* by: столбец (или список столбцов), по которому происходит сортировка.
* axis: ось, по которой производится сортировка (0 для строк, 1 для столбцов).
* ascending: порядок сортировки (True для возрастания, False для убывания).
* inplace: следует ли изменять оригинальный DataFrame (True) или создавать новый DataFrame (False).
* kind: метод сортировки (по умолчанию 'quicksort').
* na_position: куда поставить значения NaN ('last' или 'first').
* ignore_index: сбросить индексы после сортировки (True) или сохранить (False).
* key: передать функцию для применения к значениям перед сортировкой.

````python
df_amazon.sort_values(by='Year')
````

### Задание №5*

**Условие**: Выведите список (в виде объекта list) уникальных авторов за 2011 год, чье количество отзывовов хотя бы на одну книгу было больше 75 квантиля.

Для начала отсортирум авторов за 2011 год при помощи `df_amazon['Year'] == 2011`:

````python
df_2011 = df_amazon[df_amazon['Year'] == 2011]
df_2011
````

Определим значение 75-го квантиля:

````python
quantile_75 = df_2011['Reviews'].quantile(0.75)
quantile_75
````

Теперь посчитаем количество отзывов для каждого из выбранных авторов. Для этого воспользуемся методом `groupby('Author')['Reviews']`, который разобьёт датафрейм по уникальным значениям в столбце Author. Появятся группы, где каждая подгруппа будет относится к одному и тому же автору. А метод `.size()` позволит считывать количество отзывов для каждой книги этого автора

````python
author_reviews_counts = df_2011[df_2011['Reviews'] > quantile_75].groupby('Author').size()
author_reviews_counts
````

Наконец, создадим список, содержащий имена авторов, которые соответствуют заданному условию, т. е. список авторов, у котрых хотя бы одна книга имеет количество отзывов больше 75-го квантиля. Чтобы сделать это обратимся к индексу объекта `author_reviews_counts` (объект, который содержит количество книг для каждого авотра, соответствующего условию) и получим авторов, которые нам необходимы. После этого преобразуем полученный результат в список (метод `.tolist()`) и выведем ответ:

````python
selected_authors = author_reviews_counts.index.tolist()
selected_authors
````

### Задание №6

**Условие**: Создайте столбец genre_categories содержащий данные из столбца Genre, преобразованным в тип category, а затем создайте столбец genre_codes, содержащий коды категорий.

Создадим столбец category, котрый будет являться копией столбца Genre (воспользуемся методом `.astype(Genre)`):

````python
# Создание столбца genre_categories
df_amazon['genre_categories'] = df_amazon['Genre'].astype('category')
````

Воспользуемся такой особенностью pndas, которая позволяет преобразовывать столбцы в тип данных "категория" и выполнять над ним ряд специальных операций. Примеры таких операций:
1. `cat.codes`: Возвращает коды категорий (целочисленные индексы).
2. `cat.categories`: Возвращает уникальные категории.
3. `cat.add_categories(new_categories)`: Добавляет новые категории.
4. `cat.remove_categories(removed_categories)`: Удаляет указанные категории.
5. `cat.reorder_categories(new_categories)`: Переупорядочивает категории.
6. `cat.as_ordered()`: Преобразует категории в упорядоченные.
7. `cat.as_unordered()`: Преобразует категории в неупорядоченные.
8. `cat.rename_categories(new_categories)`: Переименовывает категории.
9. `cat.ordered`: Возвращает True, если категории упорядочены.
10. `cat.add_ordered(new_categories)`: Добавляет упорядоченные категории.

В данном случае будет удобно использовать `cat.codes`, что позволит создать столбец genre_codes, содержащий целочисленные коды для каждой категории в столбце genre_categories. Коды задаются автоматически, и они представляют уникальные идентификаторы для каждой категории. В данном случае, Fiction получил код 0, Non Fiction - код 1, Science - код 2.

````python
# Создание столбца genre_codes
df_amazon['genre_codes'] = df_amazon['genre_categories'].cat.codes
````

Наконец, создадим для удобства новый DataFrame, который будет состоять из 3 столбцов: Genre (для сравнения с полученным результатом), genre_categories и genre_codes.

````python
df_new = df_amazon[['Genre', 'genre_categories', 'genre_codes']]
df_new
````

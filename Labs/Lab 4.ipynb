## Задание №2

**Условие**: Приведите все названия столбцов к нижнему регистру и замените пробелы на знак '_'.


Приведём все названия столбцов к нижнему регистру, воспользовавшись методом `str.lower()`, применим его к столбцам, получим: `df.columns.str.lower()`. Осталось заменить все символы пробелов на '_', для этого воспользуемся методом `str.replace(' ', '_')`. Таким образом, записав все мотоды в одну строку, получим:

df.columns = df.columns.str.lower().str.replace(' ', '_')

Заметим, что в этом случае столбец с индексом не привёлся к нижнему регистру, поэтому приведём его отдельно:

````python
# Приведение индекса к нижнему регистру
df.index.name = df.index.name.lower()

# Вывод первых нескольких строк для проверки
df.head()
````

## Задание №4

**Условие**: Отобразите в виде столбчатой диаграммы страны ('geography') по среднему балансу счетов клиентов банка ('balance').

### 1. Выбор данных

Для начала, сгруппируем данные в датафрейме `df` по столбцу `geography` (страна) с помощью метода `groupby('geography')` и выведем средний баланс для каждой страны в столбце 'balance'. Результат представляет собой новый датафрейм с двумя столбцами: 'geography' и 'balance'.:

````python
# Группировка данных по странам и вычисление среднего баланса
average_balance_by_country = df.groupby('geography')['balance'].mean()
average_balance_by_country
````

### 2. Сортировка результатов

Отсортируем результат в порядке убывания среднего баланса с помощбю метода `.sort_values(ascending=False)`. Таким образом, получим сгруппированные и отсортированные данные:

````python
# Группировка данных по странам и вычисление среднего баланса
average_balance_by_country = df.groupby('geography')['balance'].mean().sort_values(ascending=False)
average_balance_by_country
````

### 3. Создадим диаграмму, где отобразим полученные данные

Создадим новую фигуру для графика и зададим ее размеры (ширина 10, высота 6):

```
plt.figure(figsize=(10, 6))
plt.show()
```
Создадим столбчатую диаграмму (параметр `kind='bar'`), зададим также параметр цвета (`color='skyblue'`). Добавим названия для каждой из осей, а также название самой таблицы:
```
plt.title('Средний баланс счетов клиентов по странам')
plt.xlabel('Страна')
plt.ylabel('Средний баланс')
```
Таким образом, код для создадния и вывода диаграммы будет выглядеть следующим образом:

````python
plt.figure(figsize=(10, 6))
average_balance_by_country.plot(kind='bar', color='skyblue')
plt.title('Средний баланс счетов клиентов по странам')
plt.xlabel('Страна')
plt.ylabel('Средний баланс')
plt.show()
````

### Задание №6*

**Условие**: Постройте столбчатую диаграмму с накоплением для общего количества продуктов по различным странам и полу клиентов банка. Укажите легенду, чтобы было легче прочитать график.

### 1. Фильтрация данных
Сгруппируем данные по полу и стране:
```
df.groupby(['geography', 'gender'])
```
Тут же подсчитаем общее количество продуктов для каждой из полученных групп:

````python
products_by_country_gender = df.groupby(['geography', 'gender'])['numofproducts'].sum()
products_by_country_gender
````

Воспользуемся методом `unstack()` для преобразования уровней индексов в столбцы, чтобы было удобнее работать с данными. Таким образом, получим датафрейм только с необходимыми нам для работы данными:

````python
products_by_country_gender = df.groupby(['geography', 'gender'])['numofproducts'].sum().unstack()
products_by_country_gender
````

### 2. Создание диаграммы

Аналогично предыдущему заданию, укажем размер диаграммы, оформим её, задав цвета для столбцов в соответствии с полом человека и присвоим названия осям и самой диаграмме.

*Будем использовать библиотеку seaborn, чтобы иметь болеее богатую палитру (захотел задать красивые цвета стобцам, а также просто посмотреть какие есть библиотеки).

````python
import seaborn as sns

# Группировка данных по странам и полу и подсчет общего количества продуктов
products_by_country_gender = df.groupby(['geography', 'gender'])['numofproducts'].sum().unstack()

# Построение столбчатой диаграммы с накоплением с использованием seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x=products_by_country_gender.index, y=products_by_country_gender['Female'], label='Female', color='lightcoral')
sns.barplot(x=products_by_country_gender.index, y=products_by_country_gender['Male'], bottom=products_by_country_gender['Female'], label='Male', color='lightblue')

# Добавление легенды
plt.legend(title='Gender', loc='upper right')

# Настройка осей и заголовка
plt.title('Общее количество продуктов по странам и полу клиентов банка')
plt.xlabel('Страна')
plt.ylabel('Общее количество продуктов')

# Отображение графика
plt.show()
````
